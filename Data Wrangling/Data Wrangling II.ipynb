# DUPLICATE HANDLING AND OUTLIERS

```python
# Removing duplicate rows

dup = pd.DataFrame(data)
duplicates = dup.duplicated()
duplicates
df = df.drop_duplicates()
print("\nDataFrame after removing duplicates:\n", df)
```

```python
# HANDLING OUTLIERS USING IQR METHOD
df = pd.DataFrame(data)
def handle_outliers(df, column):
    # First defining the quantiles
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lb = Q1 - 1.5 * IQR #Lower Bound
    ub = Q3 + 1.5 * IQR #Upper Bound
    print(f"{column} - IQR: {IQR}, Lower bound: {lb}, Upper bound: {ub}")
    
    # Detect outliers
    outliers = df[(df[column] < lb) | (df[column] > ub)]
    print(f"Outliers in {column}:\n", outliers)

    # Remove outliers
    df_filtered = df[~((df[column] < lb) | (df[column] > ub))]
    return df_filtered

# Handle outliers
df_no_outliers = handle_outliers(df, 'Fare')
df_no_outliers = handle_outliers(df_no_outliers, 'Age')

print("\nDataFrame after handling outliers:\n", df_no_outliers)
```
