# HANDLE MISSING VALUES

```python
# 1. Identify the columns with missing values
dataframe.isnull().any() # isnull().any() only tells if there is any missing value present it return True else False
dataframe.isnull().sum() # sum() will count the no of missing values in each column and return the number
```

```python
# 2. Use MEAN imputation for the AGE column
mean = dataframe['Age'].mean()
Mean_imputation = dataframe['Age'] = dataframe['Age'].fillna(mean)
Mean_imputation
```

```python
# Checking for null data after Mean imputation in Age Column
dataframe.isnull().sum()
```

```python
# 2. Use MODE imputation for the Embarked column
mode = dataframe['Embarked'].mode()[0]
dataframe['Embarked'] = dataframe['Embarked'].fillna(mode)
```

```python
# Checking for null data after Mode imputation in Embarked Column
dataframe.isnull().sum()
```

```python
dataframe = dataframe.drop(columns=['Cabin'])
```

```python
# Checking for null data after droping the Cabin column due to high percentage of null values
dataframe.isnull().sum()
```

# DATA TYPE CONVERSION

```python
dataframe.dtypes # Before changing datatype
```

```python
dataframe['Pclass'] = dataframe['Pclass'].astype('category')
```

```python
dataframe.dtypes # After changing the datatype of Pclass
```

# NORMALIZATION AND SCALING

FORMULA (value - min) / (max - min)

```python
dataframe['Fare'] = (dataframe['Fare'] - dataframe['Fare'].min())/(dataframe['Fare'].max() - dataframe['Fare'].min())
dataframe['Fare']
```

# Encoding Categorical Variables
ONE-HOT ENCODE

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame(dataframe)
ohe = OneHotEncoder(drop='first', sparse_output=False)

# Apply OneHotEncoder to 'Sex' and 'Embarked' columns
encoded_columns = ohe.fit_transform(df[['Sex', 'Embarked']])
encoded_df = pd.DataFrame(encoded_columns, columns=ohe.get_feature_names_out(['Sex', 'Embarked']))

df = pd.concat([df.drop(columns=['Sex', 'Embarked']), encoded_df], axis=1)

df
```
