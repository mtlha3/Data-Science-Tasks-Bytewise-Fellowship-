```python
import pandas as pd
import numpy as np
```

```python
file_path = 'C:/Users/Asennus/Bytewise Fellowship/WorldBank.csv'
df = pd.read_csv(file_path)
```

```python
#Calculating mean, median And standard deviation

variable = 'GDP per capita (USD)'
mean_value = df[variable].mean()
median_value = df[variable].median()
std_deviation = df[variable].std()

mean_value, median_value, std_deviation
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.boxplot(df[variable])
plt.title(f'Box Plot of {variable}')
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(df[variable], bins=30, kde=True)
plt.title(f'Histogram of {variable}')
plt.show()
```

```python
#Interpret Skewness and Kurtosis

from scipy.stats import skew, kurtosis

# Calculate skewness and kurtosis
skewness_value = skew(df[variable].dropna())
kurtosis_value = kurtosis(df[variable].dropna())

skewness_value, kurtosis_value
```

```python
#Formulate Null and Alternative Hypotheses


#For example, let's test if the mean GDP per capita is significantly different from a hypothetical value, say $10,000.
#Null hypothesis (H0): The mean GDP per capita is $10,000.
#Alternative hypothesis (H1): The mean GDP per capita is not $10,000
```

```python
#Conduct a t-test

from scipy.stats import ttest_1samp
hypothetical_mean = 8000
t_stat, p_value = ttest_1samp(df[variable].dropna(), hypothetical_mean)
t_stat, p_value
```

```python
#Interpret the p-value and Conclude

alpha = 0.05 
if p_value < alpha:
    print(f"Reject the null hypothesis. The mean GDP per capita is significantly different from ${hypothetical_mean}.")
else:
    print(f"Fail to reject the null hypothesis. There is no significant difference in the mean GDP per capita from ${hypothetical_mean}.")
```
