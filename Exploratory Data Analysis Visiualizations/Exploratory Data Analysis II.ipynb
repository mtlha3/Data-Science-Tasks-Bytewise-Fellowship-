# EXPLORATORY DATA ANALYSIS II
Categorical Variable Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:/Users/Asennus/Bytewise Fellowship/Assignment 5/Exploratory Data Analysis/train.csv')
data.head()
```

```python
data['Title'] = data['Name'].str.extract(r'([A-Za-z]+)\.')
data
```

# BAR PLOT

```python
# Bar Plot
categorical_features = ['Pclass', 'Sex', 'Embarked', 'Title']
    
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=feature, data=data)
    plt.title(f'Bar Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

# VIOLIN PLOT

```python
# Violin Plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='Pclass', y='Fare', data=data)
plt.title('Violin Plot b/w Pclass and Fare')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```

#  Plot swarm plots

```python
# Swarm plot
#for surived vs Age
plt.figure(figsize=(12, 6))
sns.swarmplot(x='Survived', y='Age', data=data)
plt.title('Swarm Plot b/w Survived and Age')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

#For Survived vs Fare
plt.figure(figsize=(12, 6))
sns.swarmplot(x='Survived', y='Fare', data=data)
plt.title('Swarm Plot b/w Survived and Fare')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()
```

# FEATURE INTERACTIONS

```python
# Converting into numeric data because aplha cannot be plot into to graphs
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
data['Embarked'] = data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# Selecting only numeric columns for pair plotting because of conversion error
numeric_df = data.select_dtypes(include=['float64', 'int64'])

sns.pairplot(numeric_df)
plt.suptitle('Pair Plot of Numerical Features', y=1.02)
plt.show()
```
